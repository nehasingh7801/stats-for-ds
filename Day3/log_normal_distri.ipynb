```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('income.csv')
df.head()
```

```python
g = sns.barplot(x='income($)', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income($)',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
g.set(xscale="log");
```
